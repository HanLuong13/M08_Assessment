```python
#imports
import pandas as pd
import seaborn as sns
import requests
import matplotlib.pyplot as plt

#api
urls = ['https://api.census.gov/data/2018/abscs?get=NAME,GEO_ID,NAICS2017_LABEL,SEX,ETH_GROUP,RACE_GROUP,RACE_GROUP_LABEL,VET_GROUP,FIRMPDEMP&for=state:*&NAICS2017=00&key=f0cb70159f471ac5cc9b56ba7b7661051d8cfe75',
    'https://api.census.gov/data/2018/abscb?get=NAME,GEO_ID,NAICS2017_LABEL,SEX,ETH_GROUP,RACE_GROUP,VET_GROUP,FIRMPDEMP&for=state:*&NAICS2017=00&BUSCHAR=A1&QDESC=B01&key=f0cb70159f471ac5cc9b56ba7b7661051d8cfe75',
    'https://api.census.gov/data/2018/abscbo?get=NAME,GEO_ID,NAICS2017_LABEL,OWNER_SEX,OWNER_SEX_LABEL,OWNER_ETH,OWNER_ETH_LABEL,OWNER_RACE,OWNER_RACE_LABEL,OWNER_VET,OWNPDEMP&for=state:*&OWNCHAR=CG&NAICS2017=00&QDESC=O02&key=f0cb70159f471ac5cc9b56ba7b7661051d8cfe75',
    'https://api.census.gov/data/2018/abstcb?get=NAME,GEO_ID,NAICS2017_LABEL,TECHUSE,FIRMPDEMP&for=state:*&NAICS2017=00&key=f0cb70159f471ac5cc9b56ba7b7661051d8cfe75']
```

```python
#function to extract data
def getDF(url):
    response = requests.get(url).json()
    labels = response[0]
    rows = response[1:]
    df = pd.DataFrame(rows, columns=labels)
    return df
```

```python
compSummary = getDF(urls[0])
#bizChar = getDF(urls[1])
#bizOwner = getDF(urls[2])
#techChar = getDF(urls[3])
```

```python
compSummary.shape
```

```python
totalRace = compSummary[['NAME','RACE_GROUP_LABEL']]
groupRace = totalRace.groupby(['RACE_GROUP_LABEL']).count()
groupRace = groupRace.sort_values(by='NAME',ascending=False)
groupRace = groupRace[1:11]
groupRace.plot.bar(title = 'Count of Companies per Racial Group');
```
